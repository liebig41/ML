## 1.3学会画条形图

### 横着的，竖着的！？？

```python
a = ['战狼', '速度与激情', '功夫瑜伽', '西游伏妖篇', '变形金刚', '摔跤吧', '加勒比海']
b = [56.01, 23.94, 17.53, 16.49, 15.45, 12.96, 11.8]

plt.figure(figsize=(20,8),dpi=150)

plt.barh(range(len(a)), b, height=0.3)

plt.yticks(range(len(a)),a,rotation=0)

plt.xlabel('电影名')
plt.ylabel('票房')
plt.title('近期电影票房一览')

plt.grid()

plt.show()
```

```python
# 提供示例数据
x_values = [1, 2, 3, 4, 5]
y_values = [10, 15, 7, 20, 12]

# 绘制折线图
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='折线图')

# 添加标题和标签
plt.title('简单折线图示例')
plt.xlabel('X 轴标签')
plt.ylabel('Y 轴标签')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```

```python
# 提供示例数据
x_values = [1, 2, 3, 4, 5]
y_values = [10, 15, 7, 20, 12]

# 绘制散点图
plt.scatter(x_values, y_values, color='r', marker='o', label='散点图')

# 添加标题和标签
plt.title('简单散点图示例')
plt.xlabel('X 轴标签')
plt.ylabel('Y 轴标签')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

# 提供示例数据
data = [1, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5, 6, 6, 6, 7, 8, 8, 8, 9, 9, 10]

# 绘制直方图
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# 添加标题和标签
plt.title('简单直方图示例')
plt.xlabel('数据')
plt.ylabel('频率')

# 显示图形
plt.show()
```
